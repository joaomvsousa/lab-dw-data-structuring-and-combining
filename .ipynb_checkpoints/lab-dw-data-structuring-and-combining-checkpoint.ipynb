```python
def lower_names_spaces(insurance):
    insurance.columns = insurance.columns.str.lower()
    insurance.columns = insurance.columns.str.replace(" ", "_")
    return insurance
```

```python
# cleaning column state, first replacing short names and after put everything uppercase
insurance["state"] = insurance["state"].replace("WA", "WASHINGTON")
insurance["state"] = insurance["state"].replace("AZ", "ARIZONA")
insurance["state"] = insurance["state"].replace("CALI", "CALIFORNIA")
insurance["state"] = insurance["state"].str.upper()
```

```python
#cleaning column gender, transforming all data who isn't M or F in it. 
insurance["gender"] = insurance["gender"].replace("Male", "M")
insurance["gender"] = insurance["gender"].replace("female", "F")
insurance["gender"] = insurance["gender"].replace("Femal", "F")
```

```python
def extract_middle_number(string):
    if isinstance(string, str) and string.count('/') == 2:  # Check if the value is a string and contains two slashes
        parts = string.split('/')  # Split the string by '/'
        if len(parts) == 3:  # Check if the string has three parts
            middle_part = parts[1]  # Get the middle part
            if middle_part.isdigit():  # Check if the middle part is composed of digits
                return middle_part  # Return the middle part
    return string  # Return the original value if the conditions are not met
```

```python
insurance['number_of_open_complaints'] = insurance['number_of_open_complaints'].apply(extract_middle_number)
```

```python
insurance['number_of_open_complaints'].value_counts()
```

```python
insurance["vehicle_class"] = insurance["vehicle_class"].replace("Sports Car", "Luxury")
insurance["vehicle_class"] = insurance["vehicle_class"].replace("Luxury SUV", "Luxury")
insurance["vehicle_class"] = insurance["vehicle_class"].replace("Luxury Car", "Luxury")
```

```python
def drop_rows_with_many_missing_values(insurance, threshold=8):
    missing_counts = insurance.isnull().sum(axis=1)
    filtered_df = insurance[missing_counts <= threshold]
    return filtered_df
```
